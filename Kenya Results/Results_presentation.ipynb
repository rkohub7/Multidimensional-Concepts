```python
import matplotlib.pyplot as plt

# Data for plotting
indicators = [
    "Stunting", "Health/Vaccination Card", "Underweight",
    "Information Devices", "Water Source", "Toilet Facility",
    "Dwelling Material", "Indoor Air Pollution"
]
percentages = [18.13, 13.99, 12.58, 36.07, 45.23, 70.06, 55.43, 85.20]

# Normalize percentages to scale circle sizes
max_size = 3000  # Max size for the largest circle
sizes = [max_size * (p / 100) for p in percentages]

# Plotting the circles
fig, ax = plt.subplots()
ax.scatter(indicators, [1] * len(indicators), s=sizes, alpha=0.5)

# Formatting the plot
ax.set_xlim(-1, len(indicators))
ax.set_ylim(0.5, 1.5)
ax.set_xticks(range(len(indicators)))
ax.set_xticklabels(indicators, rotation=45, ha='right')
ax.set_yticks([])  # Hide y-axis
ax.set_title("Proportional Circles for Deprivation Indicators")

plt.show()
```

```python
import pandas as pd

# Create a DataFrame for better visualization
data = {
    'Indicators': indicators,
    'Deprived (%)': percentages
}
df = pd.DataFrame(data)

# Plotting a bar chart
plt.figure(figsize=(10, 6))
plt.bar(df['Indicators'], df['Deprived (%)'], color='skyblue', edgecolor='black')

# Adding title and labels
plt.title('Deprivation Indicators Percentage', fontsize=16)
plt.xlabel('Indicators', fontsize=12)
plt.ylabel('Deprived (%)', fontsize=12)

# Rotating the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the bar chart
plt.show()
```

```python
# Plotting a bar chart with labels
plt.figure(figsize=(10, 6))
bars = plt.bar(df['Indicators'], df['Deprived (%)'], color='skyblue', edgecolor='black')

# Adding title and labels
plt.title('Deprivation Indicators Percentage', fontsize=16)
plt.xlabel('Indicators', fontsize=15)
plt.ylabel('Deprived (%)', fontsize=12)

# Adding value labels on top of the bars
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 1, round(yval, 2), ha='center', va='bottom')

# Rotating the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the bar chart
plt.show()
```

```python
# Updated data for stack diagram
updated_percentages = [7.70, 2.60, 2.70, 15.70, 19.10, 25.90, 11.60, 14.70]

# Plotting the stacked bar chart
plt.figure(figsize=(10, 6))
plt.barh("Deprivation Indicators", updated_percentages, color='skyblue', edgecolor='black')

# Adding the percentage labels
cumulative = 0
for i, (indicator, percentage) in enumerate(zip(indicators, updated_percentages)):
    plt.text(cumulative + percentage / 2, 0, f'{indicator} ({percentage}%)', ha='center', va='center')
    cumulative += percentage

# Adding title and labels
plt.title('Stacked Diagram of Deprivation Indicators Summing to 100%', fontsize=16)
plt.xlabel('Percentage (%)', fontsize=12)
plt.yticks([])  # Hide y-axis labels

# Display the stacked bar chart
#plt.show()
```

```python
# Plotting a pie chart with different colors
plt.figure(figsize=(8, 8))
colors = plt.cm.Paired(range(len(indicators)))  # Use a colormap to get different colors

plt.pie(
    updated_percentages, 
    labels=indicators, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=colors
)

# Adding title to the pie chart
plt.title('Contribution to MPI', fontsize=16)

# Display the pie chart
plt.show()
```

```python
# Data for poverty incidence by residence
residence = ['Rural', 'Urban']
poverty_incidence = [79.84, 42.19]

# Plotting the bar chart for poverty incidence by residence
plt.figure(figsize=(8, 6))
plt.bar(residence, poverty_incidence, color=['forestgreen', 'royalblue'], edgecolor='black')

# Adding title and labels
plt.title('Poverty Incidence by Residence', fontsize=16)
plt.xlabel('Residence', fontsize=12)
plt.ylabel('Poverty Incidence (%)', fontsize=12)

# Adding value labels on top of the bars
for i, v in enumerate(poverty_incidence):
    plt.text(i, v + 2, f'{v:.2f}%', ha='center', va='bottom')

# Display the bar chart
plt.show()
```
